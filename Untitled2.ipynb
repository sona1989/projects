```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = r'C:\Users\manish\Desktop\data science\DATA.csv'
df= pd.read_csv(data)
```

```python
#drop 2 columns as it has no relevant with the data 
df.drop(['UnListed Date'], axis = 1, inplace = True)
df.drop(['UnListed By'], axis = 1, inplace = True)
df.drop('ID', axis = 1, inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
cat_features= [column_name for column_name in df.columns if df[column_name].dtype == 'O']
df[cat_features].isnull().sum()

null_value = [feature for feature in cat_features if df[feature].isnull().sum()]
for each_feature in null_value:
    mode_val=df[each_feature].mode()[0]
    df[each_feature].fillna(mode_val,inplace=True)
```

```python
df.head()
```

```python
def encode_data(feature_name):

    ''' 

    This function takes feature name as a parameter and returns mapping dictionary to replace(or map) categorical data with numerical data.

    '''

    mapping_dict = {}

    unique_values = list(df[feature_name].unique())

    for idx in range(len(unique_values)):

        mapping_dict[unique_values[idx]] = idx

    return mapping_dict

df['Make'].replace(encode_data('Make'),inplace = True)

df['Model'].replace(encode_data('Model'),inplace = True)

df['Ram'].replace(encode_data('Ram'),inplace = True)

df['Rom'].replace(encode_data('Rom'),inplace = True)
df['Color'].replace(encode_data('Color'),inplace = True)
df['Orignal Kit'].replace(encode_data('Orignal Kit'),inplace = True)
df['Vendor Name'].replace(encode_data('Vendor Name'),inplace = True)
df['Mobex Grade'].replace(encode_data('Mobex Grade'),inplace = True)
df['Status'].replace(encode_data('Status'),inplace = True)
df['Entry By'].replace(encode_data('Entry By'),inplace = True)
df['Reject Reason'].replace(encode_data('Reject Reason'),inplace = True)
df['ASM'].replace(encode_data('ASM'),inplace = True)
df['Area'].replace(encode_data('Area'),inplace = True)
```
